```python
import pandas as pd
# import numpy as np
import os
print(os.getcwd())
```

```python
# Charger le dataset
file_path = "./Dataset/HDHI Admission data.csv" 
data = pd.read_csv(file_path)

# 1. Afficher les informations de base
print("Dataset Info:")
print(data.info())

# 2. Afficher les premières lignes
print("\nPremières lignes du dataset:")
print(data.head())
```

```python
# 3. Identifier les valeurs manquantes
print("\nValeurs manquantes par colonne:")
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# 4. Convertir les colonnes de dates au format datetime
data['D.O.A'] = pd.to_datetime(data['D.O.A'], errors='coerce')
data['D.O.D'] = pd.to_datetime(data['D.O.D'], errors='coerce')

# 5. Calculer la durée de séjour (D.O.D - D.O.A)
data['STAY_DURATION'] = (data['D.O.D'] - data['D.O.A']).dt.days

# 6. Créer des colonnes dérivées
data['ADMISSION_MONTH'] = data['D.O.A'].dt.month
data['ADMISSION_YEAR'] = data['D.O.A'].dt.year
data['ADMISSION_SEASON'] = data['D.O.A'].dt.month % 12 // 3 + 1  # 1: Winter, 2: Spring, etc.

# 7. Afficher les modifications
print("\nAperçu des colonnes modifiées:")
print(data[['D.O.A', 'D.O.D', 'STAY_DURATION', 'ADMISSION_MONTH', 'ADMISSION_YEAR', 'ADMISSION_SEASON']].head())

# 8. Sauvegarder le dataset nettoyé (si nécessaire)
cleaned_file_path = "cleaned_HDHI_Admission_data.csv"
data.to_csv(cleaned_file_path, index=False)
print(f"\nDataset nettoyé sauvegardé sous: {cleaned_file_path}")
```
