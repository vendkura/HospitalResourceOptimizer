```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Charger le dataset
file_path = "./Dataset/cleaned_HDHI_Admission_data.csv" 
data = pd.read_csv(file_path)

# Statistiques descriptives des colonnes numériques
print("Statistiques descriptives :")
print(data.describe())

# Statistiques pour des colonnes spécifiques
print("\nDistribution des types d'admissions :")
print(data['TYPE OF ADMISSION-EMERGENCY/OPD'].value_counts())

print("\nDistribution des genres :")
print(data['GENDER'].value_counts())

print("\nRépartition des âges :")
print(data['AGE'].describe())
```

## Visualisations


```python
# Histogramme de la durée de séjour
plt.figure(figsize=(8, 5))
data['STAY_DURATION'].hist(bins=30, edgecolor='black')
plt.title("Distribution de la durée de séjour")
plt.xlabel("Durée de séjour (jours)")
plt.ylabel("Nombre de patients")
plt.show()
```

## filtrage des donnes aberrantes car nous avons des durees de -0 

```python
# Filtrer les valeurs aberrantes pour la durée de séjour
print(f"Taille initiale du dataset :{data.shape}")
data = data[(data['STAY_DURATION'] >= 0) & (data['STAY_DURATION'] <=100)]

# Filtrer les valeurs aberrantes pour l'âge
data = data[(data['AGE']>0) & (data['AGE']<=110)]

# Verification de la taille du dataset apres filtrage
print(f"Taille apres filtrage :{data.shape}")

# Stats descriptive après filtrage
print(data.describe())
```
